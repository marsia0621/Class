```python
html = urlopen("http://www.mnet.com/chart/TOP100/")
html
```

```python
bsObject = BeautifulSoup(html, "html.parser")
bsObject
```

```python
dt = bsObject.find("tbody")
```

```python
print(dt)
```

```python
title = dt.select('a')[2].get_text()
title
```

```python
artist = dt.select('a')[3].get_text()
artist
```

```python
album = dt.select('a')[4].get_text()
album
```

```python
dt = bsObject.tbody.find_all('tr')[0]
dt
```

```python
for n in range (0,50):
    dt = bsObject.tbody.find_all('tr')[n]
    title = dt.select('a')[2].get_text()
    artist = dt.select('a')[3].get_text()
    album = dt.select('a')[4].get_text()
    print(n+1, title, artist, album)
```

```python
with open('mnet.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['rank', 'title', 'artist', 'album'])
        
    for n in range (0,50):
        dt = bsObject.tbody.find_all('tr')[n]
        title = dt.select('a')[2].get_text()
        artist = dt.select('a')[3].get_text()
        album = dt.select('a')[4].get_text()
        writer.writerow([n+1, title, artist, album])
```
